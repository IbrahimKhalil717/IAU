### Task 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('new_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((7,7), np.float32) / 49
smoothed = cv2.filter2D(img, -1, kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(smoothed)
plt.title("7x7 Box Filter")
plt.axis("off")

plt.show()
```

### Task 2

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('new_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gauss5 = cv2.GaussianBlur(img, (5,5), 0)
gauss21 = cv2.GaussianBlur(img, (21,21), 0)

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gauss5)
plt.title("5x5 Gaussian")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(gauss21)
plt.title("21x21 Gaussian")
plt.axis("off")

plt.show()
```

### Task 3

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('new_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_float = img.astype(np.float32) / 255.0

laplacian = cv2.Laplacian(img_float, cv2.CV_32F, ksize=3)

sharpened = np.clip(img_float - laplacian, 0, 1)

plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(img_float)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(laplacian)
plt.title("Laplacian")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sharpened)
plt.title("Sharpened")
plt.axis("off")

plt.show()
```
