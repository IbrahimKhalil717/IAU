### Task 1

```python
from skimage import data, exposure
import matplotlib.pyplot as plt
import numpy as np


image = data.moon()


p3, p80 = np.percentile(image, (3, 80))


rescaled = exposure.rescale_intensity(image, in_range=(p3, p80))


plt.figure()
plt.imshow(rescaled, cmap='gray')
plt.title('Rescaled Moon Image')
plt.axis('off')

plt.figure()
plt.hist(rescaled.ravel(), bins=256)
plt.title('Histogram after Rescaling')
plt.show()
```

### Task 2

```python
from skimage import data, exposure
import matplotlib.pyplot as plt

image = data.moon()

equalized = exposure.equalize_hist(image)

plt.figure()
plt.imshow(equalized, cmap='gray')
plt.title('Histogram Equalized Moon Image')
plt.axis('off')

plt.figure()
plt.hist(equalized.ravel(), bins=256)
plt.title('Histogram after Equalization')
plt.show()
```

### Task 3

```python
from skimage import data, exposure
import matplotlib.pyplot as plt

reference = data.rocket()
image = data.chelsea()

matched = exposure.match_histograms(image, reference, channel_axis=-1)

plt.figure()
plt.imshow(image)
plt.title("Original chelsea")
plt.axis("off")

plt.figure()
plt.imshow(reference)
plt.title("Reference rocket")
plt.axis("off")

plt.figure()
plt.imshow(matched)
plt.title("Histogram Matched chelsea and rocket")
plt.axis("off")

plt.figure()
plt.hist(image[..., 0].ravel(), bins=256)
plt.title("Red channel histogram: original chelsea")
plt.show()

plt.figure()
plt.hist(matched[..., 0].ravel(), bins=256)
plt.title("Red channel histogram: matched")
plt.show()
```
